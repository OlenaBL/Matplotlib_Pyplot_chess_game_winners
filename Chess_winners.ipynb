Count for winners in chess games using Matplotlib Pyplot 

```python
def chess_winners(players):
    num_victories = {}
    for player in players:
        if player in num_victories:
            num_victories[player] += 1
        else:
            num_victories[player] = 1
    return num_victories
players = ["Ada", "Jerry", "Tom", "Bobby", "Bobby", "Bobby", "Ada", "Ada", "Bobby", "Bobby", "Tom"]
num_victories = chess_winners(players)
print(num_victories)

#winners = {}
for player, victories in num_victories.items():
    print(player, victories)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
player = [player for player in num_victories.keys()]
victories = [victories for victories in num_victories.values()]
plt.plot(player, victories, "go-", linewidth=2, markersize=14, linestyle='dotted')
plt.title("Game winners")
plt.show()
```
